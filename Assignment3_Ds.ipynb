```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv("data.csv", encoding='latin1')
data.head()
```

```python
data.columns
```

```python
x = data['no2']
```

```python
x = pd.to_numeric(x, errors='coerce')
x = x.dropna()
```

```python
r = 102303116

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)

z = x + ar * np.sin(br * x)
```

```python
mu = np.mean(z)
```

```python
variance = np.var(z)
```

```python
lambda_ = 1 / (2 * variance)
```

```python
c = 1 / math.sqrt(2 * math.pi * variance)
```

```python
print("mu =", mu)
print("lambda =", lambda_)
print("c =", c)
```
